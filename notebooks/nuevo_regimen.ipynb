```r
source("../src/utils/nuevo_regimen.R")
mujeres <- read.csv("../data/input/mujeres.csv")
```

```r
nuevo_regimen(mujeres, 57, 0.0788, 0.03, 0.132, 1423500, 1423500, 25, 2.3)
```

```r
# Parámetros constantes
  tabla_mortalidad <- mujeres
  r <- 0.0788
  e <- 0.03
  tc <- 0.132
  Sm <- 1423500
  h <- 2.3

  # Valores variables
  x_values <- c(57, 62, 67)
  n_values <- c(25, 35, 42)
  S_values <- 1:25 * 1423500

  # Función para crear una tabla para un valor específico de x
  crear_tabla <- function(x_val) {
    # Crear matriz para almacenar resultados
    resultados <- matrix(NA, nrow = 25, ncol = 3)
    colnames(resultados) <- paste(n_values, "años")

    # Llenar la matriz
    for(i in 1:25) {
      S <- S_values[i]
      for(j in 1:3) {
        n <- n_values[j]
        resultados[i, j] <- nuevo_regimen(tabla_mortalidad, x_val, r, e, tc, Sm, S, n, h)
      }
    }

    # Crear data frame con formato
    df <- data.frame(
      SMMLV = 1:25,
      resultados
    )

    # Formatear números a pesos colombianos
    for(col in 2:4) {
      df[, col] <- format(round(df[, col]), big.mark = ",", scientific = FALSE)
    }

    return(df)
  }

  # Crear las tres tablas
  tabla_x57 <- crear_tabla(57)
  tabla_x62 <- crear_tabla(62)
  tabla_x67 <- crear_tabla(67)

  # Mostrar las tablas con formato
  cat("╔══════════════════════════════════════════════════════════════╗\n")
  cat("║                MESADA PENSIONAL - EDAD 57 AÑOS               ║\n")
  cat("╚══════════════════════════════════════════════════════════════╝\n\n")
  print(tabla_x57, row.names = FALSE)

  cat("\n\n╔══════════════════════════════════════════════════════════════╗\n")
  cat("║                MESADA PENSIONAL - EDAD 62 AÑOS               ║\n")
  cat("╚══════════════════════════════════════════════════════════════╝\n\n")
  print(tabla_x62, row.names = FALSE)

  cat("\n\n╔══════════════════════════════════════════════════════════════╗\n")
  cat("║                MESADA PENSIONAL - EDAD 67 AÑOS               ║\n")
  cat("╚══════════════════════════════════════════════════════════════╝\n\n")
  print(tabla_x67, row.names = FALSE)
```
